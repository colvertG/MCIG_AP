**Serie de Taylor para la función $f(\mathbf{x})=x^2$ alrededor de un punto $f(\mathbf{x_0})=3$**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2

def df(x):
    return 2*x
    
def ddf(x):
    return 2
    
def taylor(x,x0):
    d = x-x0
    return f(x0)+d*df(x0)+0.5*d*ddf(x0)*d
```

```python
x = np.arange(-4,4,0.1)
y = [f(i) for i in x]
plt.plot(x,y,label='$x^2$')
plt.grid()
plt.legend()
plt.show()
```

```python
xc=3
local_range = 1
x = np.arange(xc-local_range,xc+local_range,0.01)

plt.plot(x,[f(i) for i in x],linewidth=3,label='Original')
plt.plot(x,[taylor(i,x0=xc) for i in x],color='red',label='Taylor 2da derivada')
plt.vlines(x=xc,ymin=2,ymax=f(xc+local_range),color='green')
plt.grid()
plt.legend()
plt.show()
```

**Serie de Taylor para la función $f(\mathbf{x})=x^4+x^2+1$ alrededor de un punto $f(\mathbf{x_0})=1$**

```python
def f(x):
    return x**4-3*x**2+1

def df(x):
    return 4*x**3-6*x
    
def ddf(x):
    return 12*x**2-6
    
def taylor(x,x0):
    d = x-x0
    return f(x0)+d*df(x0)+0.5*d*ddf(x0)*d
```

```python
x = np.arange(-2,2,0.01)
y = [f(i) for i in x]
plt.plot(x,y,label='$x^4+x^2+1$')
plt.grid()
plt.legend()
plt.show()
```

```python
xc=1
local_range = 1
x = np.arange(xc-local_range,xc+local_range,0.01)

plt.plot(x,[f(i) for i in x],linewidth=2.5,label='Original')
plt.plot(x,[taylor(i,x0=xc) for i in x],color='red',label='Taylor 2da derivada')
plt.vlines(x=xc,ymin=-2,ymax=f(xc+local_range),color='green')
plt.grid()
plt.legend()
plt.show()
```
